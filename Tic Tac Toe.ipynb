```python
from IPython.display import clear_output

def display_board(board):
    clear_output()
    x = len(board)
    j = 8
    for i in range(0,3):
        
        print(f'\t\t\t\t\t\t  {board[j-2]} | {board[j-1]} | {board[j]}')
        j = j - 3
        if j > 1:
            print('\t\t\t\t\t\t--------------')
    print('\n\n')

def replay():
    request = input('\n\n\t\t\t\tPlay again? Type Yes or No \n\n\t\t\t\t')
    if request == 'Yes':
        return True
    else:
        return False

def win_check(board,marker):
    if board[0] == marker:
        if board[0] == board[1] and board[1] == board[2]:
            return True
        elif board[0] == board[4] and board[4] == board[8]:
            return True
        elif board[0] == board[3] and board[3] == board[6]:
            return True
    elif board[1] == marker:
        if board[1] == board[4] and board[4] == board[7]:
            return True
    elif board[2] == marker:
        if board[2] == board[4] and board[4] == board[6]:
            return True
        elif board[2] == board[5] and board[5] == board[8]:
            return True
    elif board[3] == marker:
        if board[3] == board[4] and board[4] == board[5]:
            return True
    elif board[6] == marker:
        if board[6] == board[7] and board[7] == board[8]:
            return True
    else:
        return False


def full_board_check(board):
    x = len(board)
    flag = 0
    for i in range(0,x):
        if board[i] != 'X' and board[i] != 'O':
            flag = 1
            break
    if flag == 1:
        return False
    else:
        return True
    
def space_check(board,position):
        if board[position] != 'X' and board[position] != 'O':
            return True
        else:
            return False
        
def player_choice(board,marker):
    position = int(input('\n\n\t\t\t\t Enter the position in which you want to enter the value\n\n\t\t\t\t '))
    position = position - 1
    space = space_check(board,position)
    if space == True:
        return position
    else:
        while(space != True):
            print('\n\n\t\t\t\t That Position is not empty, choose another position\n\n\t\t\t\t ')
            position = int(input('\n\n\t\t\t\t Enter the position in which you want to enter the value\n\n\t\t\t\t '))
            position = position - 1
            space = space_check(board,position)
        return position

def place_marker(board,marker,position):
    board[position] = marker
    display_board(board)
    
def player_input():
    marker1 = input("\n\n\t\t\t\t Please pick a marker 'X' or 'O': \n\n\t\t\t\t")
    
    while marker1 != 'X' and marker1 != 'O':
        marker1 = input('\n\t\t\t\t Player 1 choose X or O: \n\t\t\t\t ')
    
    player1 = marker1
    
    print("\n\t\t\t\t Player 1 is "+player1+"\n\n\t\t\t\t Player 1's Turn ")
    marker_position = player_choice(board,player1)
    place_marker(board,player1,marker_position)
    if player1 == 'X':
        player2 = 'O'
        print("\n\t\t\t\t Player 2 is 'O' \n\n\t\t\t\t Player 2's Turn ")
        marker_position = player_choice(board,player2)
        place_marker(board,player2,marker_position)
    else:
        player2 = 'X'
        print("\n\t\t\t\t Player 2 is 'X' \n\n\t\t\t\t Player 2's Turn ")
        marker_position = player_choice(board,player2)
        place_marker(board,player2,marker_position)
    
    full = full_board_check(board)
    while full == False:
        
        print('\n\t\t\t\t Player 1')
        marker_position = player_choice(board,player1)
        place_marker(board,player1,marker_position)
        
        player1_win = win_check(board,player1)
        if player1_win == True:
            print('\n\n\t\t\t\t Congratulations!!! Player 1 has won the game')
            break
        full = full_board_check(board)
        if full == True:
            print('\n\t\t\t\t Match is tied, No more moves possible')
            break
        
        print('\n\t\t\t\t Player 2')
        marker_position = player_choice(board,player2)
        place_marker(board,player2,marker_position)
        
        player2_win = win_check(board,player2)
        if player2_win == True:
            print('\n\n\t\t\t\t Congratulations!!! Player 2 has won the game')
            break
            
        full = full_board_check(board)
    
    if player1_win == False and player2_win == False:
        print('\n\n\t\t\t\t Match is tied')
        
```
